```python
# 把rate取0-1
df["rating"] = df["rating"]/5.0

#pivot
movies = df.pivot_table(index=['userId'],columns='movieId',values='rating') # x:movie_id; y:user_id
users = df.pivot_table(index=['movieId'],columns='userId',values='rating') # x:user_id; y:movie_id
n_users, n_movies = movies.shape

df_movies = movies.fillna(0) 
df_users = users.fillna(0)
# df_movies = movies.fillna(0.5) 
# df_users = users.fillna(0.5)
```

```python
userId_list = df_users.columns
movieId_list = df_movies.columns
n_movies = len(movieId_list)
n_users = len(userId_list)
movieId_list, userId_list,n_movies, n_users
```

```python
# create a empty similarity_matrix_icf
similarity_matrix_icf = pd.DataFrame(data = np.zeros((n_movies,n_movies)), index = movieId_list, columns = movieId_list)

for movie_id in movieId_list:
    for other_movie_id in movieId_list:
        
        vec_movie = []
        vec_other_movie_id = []

        if movie_id != other_movie_id:
            
            for user_id in userId_list:

                this = df_users[user_id]
                vec_movie.append(this[movie_id])
                vec_other_movie_id.append(this[other_movie_id])
            
            similarity_matrix_icf[movie_id][other_movie_id] = np.corrcoef(np.array(vec_movie), np.array(vec_other_movie_id))[0][1]

# CSV
# similarity_matrix_icf.to_csv("similarity_matrix_v2000&600_icf.csv",header=True, sep = ',')
similarity_matrix_icf
```

```python
# 跟mid最像的100个movies
col_name = []
for i in range(1,101): # (1,101)
    col_name.append('m_id'+str(i))
n_col = len(col_name)
ll_similar_movies = pd.DataFrame(data = np.zeros((n_movies,len(col_name))), index = movieId_list, columns = col_name)
for mid in movieId_list:
    line = similarity_matrix_icf[mid].sort_values(ascending=False).index.tolist()[:n_col]
    for j in range(n_col):
        ll_similar_movies[col_name[j]][mid] = line[j]
        
# ll_similar_movies.to_csv("ll_similar_movies_100_v2000&600.csv",header=True, sep = ',')
ll_similar_movies
```

```python
# Compute similarity_matrix_ucf

# create a empty similarity_matrix_ucf
similarity_matrix_ucf = pd.DataFrame(data = np.zeros((n_users,n_users)), index = userId_list, columns = userId_list)

for user_id in userId_list:
    for other_user_id in userId_list:
        vec_user = []
        vec_other_user_id = []
        
        if user_id != other_user_id:
            
            for movie_id in movieId_list:
                
                this = df_movies[movie_id]
                vec_user.append(this[user_id])
                vec_other_user_id.append(this[other_user_id])
                    
            similarity_matrix_ucf[user_id][other_user_id] = np.corrcoef(np.array(vec_user), np.array(vec_other_user_id))[0][1]

# CSV
# similarity_matrix_ucf.to_csv("similarity_matrix_v2000&600_ucf.csv",header=True, sep = ',')
similarity_matrix_ucf
```

```python
# User_Based CF user * movie similarity matrix
avg = np.mean(df_movies, axis = 1) # each user's avg rate
sum_similarity_uid = np.sum(similarity_matrix_ucf,axis = 1) # sum[S(uid,otheruid) while otheruid in userId_list]
df_users_mean = df_users - avg

core_similarity_ucf = pd.DataFrame(data = np.zeros((n_users,n_movies)), index = userId_list, columns = movieId_list)
for mid in movieId_list:
    for uid in userId_list:
        avg_uid = avg[uid]       
        up = np.sum(df_users_mean.T[mid]*similarity_matrix_ucf[uid])
        down = sum_similarity_uid[uid]
        res = avg_uid + up/down
        core_similarity_ucf[mid][uid] = res
        
# core_similarity_ucf.to_csv("core_similarity_ucf_v2000&600.csv",header=True, sep = ',')
core_similarity_ucf
```

```python
# User_Based CF recommend

col_name = []
for i in range(1,101): 
    col_name.append('m_id'+str(i))
n_col = len(col_name)

recommend_table = pd.DataFrame(data = np.zeros((n_users,len(col_name))), index = userId_list, columns = col_name)
core_similarity_ucf_T = core_similarity_ucf.T

for uid in userId_list:
    line = core_similarity_ucf_T[uid].sort_values(ascending=False).index.tolist()[:n_col]
    for j in range(n_col):
        recommend_table[col_name[j]][uid] = line[j]

# recommend_table.to_csv("recommend_table_v2000&600.csv",header=True, sep = ',')
recommend_table
```
